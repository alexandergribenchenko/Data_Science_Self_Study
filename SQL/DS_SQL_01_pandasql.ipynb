```python
import pandas as pd
from pandasql import sqldf
```

```python
df = pd.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv")
```

```python
df.head(3)
```

```python
sqldf("""
SELECT * 
FROM df 
LIMIT 10
""")
```

```python
dict_input = { 'shop_id': [1,2,3,4,5], 
               'product_id': [1,7,3,4,5],
               'Shop_name': ['Kellogs','Fantasy Store','Laxshmi store','General Store','Corn store'],
               'quantity': [15,25,35,45,55],
               'price': [10,20,30,40,50]    
             }

df = pd.DataFrame.from_dict(dict_input)
df
```

```python
sqldf("""
WITH gato AS (SELECT shop_id,
                     product_id
			        FROM df),
     perro AS (SELECT shop_id,
                      product_id,
                      shop_name,
                      quantity
				      FROM df),
     mico AS (SELECT shop_id AS pericles,
                      product_id,
                      shop_name,
                      quantity
				      FROM df)
    

SELECT shop_id, pericles, quantity
FROM gato gt LEFT JOIN mico pr ON gt.product_id=pr.product_id  
""")
```
