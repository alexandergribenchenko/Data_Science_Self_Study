```python
import pandas as pd
```

```python
df_all = pd.read_json('example.json', orient='records', dtype='object')
df_all['sum_asks']=df_all['sum_asks'].astype('float')
df_all
```

```python
df_all.snapshot_date.max()
```

```python
df_last_date = df_all[df_all.snapshot_date==df_all.snapshot_date.max()]
df_last_date.head(20)
```

```python
df_agregada = df_last_date.groupby(['airline_code', 'rt'], as_index=False).\
           agg(max_flight_date=('flight_date', 'max'),
               min_flight_date=('flight_date', 'min')
               ).\
           sort_values(by=['airline_code', 'rt'], ascending=[True,True])

df_agregada['min_flight_date'] = pd.to_datetime(df_agregada['min_flight_date'], format='%Y%m')

df_agregada['prev_6m_flight_date'] = df_agregada['min_flight_date'] - pd.DateOffset(months=6)

df_agregada['min_flight_date'] = df_agregada['min_flight_date'].dt.strftime('%Y%m')
df_agregada['prev_6m_flight_date'] = df_agregada['prev_6m_flight_date'].dt.strftime('%Y%m')


df_agregada
```

```python
df = df_agregada.copy()
```

```python
import pandas as pd

# Supongamos que tienes un DataFrame llamado 'df' con las columnas 'airline_code', 'rt', 'max_flight_date', 'min_flight_date' y 'prev_6m_flight_date'

# Crea una lista para almacenar los nuevos registros
new_rows = []

# Itera sobre cada fila del DataFrame original
for index, row in df.iterrows():
    # Obtiene los valores de las columnas relevantes para la fila actual
    airline_code = row['airline_code']
    rt = row['rt']
    max_flight_date = row['max_flight_date']
    prev_6m_flight_date = row['prev_6m_flight_date']

    # Genera todos los meses consecutivos entre max_flight_date y prev_6m_flight_date
    current_date = pd.to_datetime(max_flight_date, format='%Y%m')  # Convierte la fecha inicial a formato de fecha
    end_date = pd.to_datetime(prev_6m_flight_date, format='%Y%m')  # Convierte la fecha final a formato de fecha

    while current_date >= end_date:
        # Crea un nuevo registro con los valores actuales
        new_row = {
            'airline_code': airline_code,
            'rt': rt,
            'flight_date': current_date.strftime('%Y%m')
        }
        new_rows.append(new_row)

        # Reduce la fecha actual en un mes
        current_date -= pd.DateOffset(months=1)

# Crea un nuevo DataFrame a partir de la lista de nuevos registros
df_to_match = pd.DataFrame(new_rows)
df_to_match
```

```python
df_join = pd.merge(df_to_match, df_all, on=['airline_code', 'rt', 'flight_date'], how='left')
df_join['snapshot_date'] = df_join['snapshot_date'].fillna('20100101')
df_join['sum_asks'] = df_join['sum_asks'].fillna(0)
df_join
```

```python
df = df_join.copy()
df['snapshot_date'] = pd.to_datetime(df['snapshot_date'], format='%Y%m%d')
indices = df.groupby(['airline_code', 'rt', 'flight_date'])['snapshot_date'].idxmax()

# Crea un nuevo DataFrame con los registros correspondientes a los Ã­ndices encontrados
new_df = df.loc[indices]

new_df = new_df.sort_values(by=['airline_code', 'rt', 'flight_date'], ascending=[True,  True, False])

new_df['snapshot_date'] = new_df['snapshot_date'].dt.strftime('%Y%m%d')

new_df = new_df.reset_index(drop=True)

new_df
```

```python
columna = 'sum_asks'
columna_categorias = ['airline_code', 'rt']
```

```python
# FunciÃ³n personalizada para contar los valores cero en la ventana mÃ³vil
def contar_ceros_ventana(serie):
    return (serie[1:] == 0).sum()

# Crear una nueva columna con el conteo de ceros en la ventana mÃ³vil por categorÃ­a
new_df['n_months_zero_asks_rw_6M'] = new_df.groupby(columna_categorias, group_keys=False)[columna].apply(lambda x: x.rolling(7).apply(contar_ceros_ventana, raw=True))

# Ajustar la nueva columna para que muestre los valores correctos
new_df['n_months_zero_asks_rw_6M'] = new_df['n_months_zero_asks_rw_6M'].shift(-6)

# Imprimir el DataFrame resultante
new_df
```
