```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crea un DataFrame de ejemplo
data = {
    'Area': ['Ventas', 'Marketing', 'Tecnología', 'Finanzas', 'Operaciones'],
    'Gerente1': [10, 15, 20, 25, 30],
    'Gerente2': [5, 10, 15, 20, 25],
    'Gerente3': [15, 20, 25, 30, 35]
}

df = pd.DataFrame(data)

# Convierte los datos a formato largo
df_long = df.melt(id_vars=['Area'], var_name='Gerente', value_name='Cantidad')

# Crea el gráfico de barras agrupadas
sns.barplot(data=df_long, x='Area', y='Cantidad', hue='Gerente')

# Ajusta el tamaño y la orientación de las etiquetas del eje x
plt.xticks(rotation=45, ha='right')

# Muestra el gráfico
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Creamos un dataframe con dos variables categóricas
data = sns.load_dataset('tips')

# Creamos un gráfico de barras apiladas
sns.countplot(x='day', hue='sex', data=data)

# Personalizamos el gráfico
plt.title('Número de comidas por día y género')
plt.xlabel('Día de la semana')
plt.ylabel('Número de comidas')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Creamos un dataframe con dos variables categóricas
data = sns.load_dataset('tips')

# Creamos una tabla de frecuencias con las dos variables categóricas
freq_table = pd.crosstab( data['sex'], data['day'])

# Creamos un heatmap
sns.heatmap(freq_table, cmap='Blues', annot=True, fmt='d')

# Personalizamos el gráfico
plt.title('Frecuencia de comidas por día y género')
plt.xlabel('Género')
plt.ylabel('Día de la semana')
plt.show()
```
