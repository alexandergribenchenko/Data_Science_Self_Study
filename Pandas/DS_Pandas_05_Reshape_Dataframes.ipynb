# Funciones para hacer reshape de dataframes

**1. Wide to long**

- pandas.melt(): Unpivot a DataFrame from wide to long format, optionally leaving identifiers set.

-pandas.wide_to_long(): is another function that can help us transform the data from wide to long. This function is less flexible but more user-friendly than melt.

- df.stack(): Return a reshaped DataFrame or Series having a multi-level index with one or more new inner-most levels compared to the current DataFrame. 


**2. Long to wide**
- df.pivot(): method allows us to reshape the data from a long to a wide format. It returns a reshaped DataFrame organized by given unique index or column values.

```python
import pandas as pd
import math
import numpy as np
```

# 01. Método df.stack() (reshape a la forma: fila , columna, valor)

```python
df = pd.DataFrame([[0, 1, 3, 4], ['mico', 'perro', 'gato', 'oso']],
                                    index=['f1', 'f2'],
                                    columns=['c1', 'c2', 'c3', 'c4'])
df
```

```python
df.stack().reset_index()
```

```python
df_stacked = df.stack().reset_index().rename(columns={'level_0':'ex_filas', 'level_1':'ex_columnas', 0:'ex_valores'})
df_stacked
```

# 02. Método df.pivot() (Inverso método df.stack()) Formato largo a formato ancho

```python
df_stacked.pivot(index='ex_filas', columns='ex_columnas', values='ex_valores')
```

```python
df = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two',
                           'two'],
                   'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'baz': [1, 2, 3, 4, 5, 6],
                   'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
df
```

```python
df.pivot(index='foo', columns='bar', values='baz')
```
