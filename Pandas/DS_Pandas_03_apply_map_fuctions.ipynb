# 01. apply Uses

## 01.01. Use: Analogue to CASE in SQL

```python
import pandas as pd
```

```python
dict_input = { 'col_01': [1,2,3,4], 
               'col_02': ['gato','perro','gato','perro'],
               'col_03': [1,5,6,7]
             }

df = pd.DataFrame.from_dict(dict_input)
df
```

```python
df['Comparador'] = df.apply(lambda x: 'iguales' if x['col_01'] == x['col_03'] 
                                                       else x['col_02'], 
                                                       axis=1)
df
```

## 01.02. Use: Concatenate multiple columns

```python
import pandas as pd
```

```python
dict_input = { 'col_01': [1,2,3,4], 
               'col_02': ['gato','perro','gato','perro'],
               'col_03': [1,5,6,7]
             }

df = pd.DataFrame.from_dict(dict_input)
df
```

```python
columnas_interes = ['col_01', 'col_02', 'col_03']
df['concatenated'] = df[columnas_interes].apply(lambda row: ''.join(row.values.astype(str)), axis=1)
df
```
