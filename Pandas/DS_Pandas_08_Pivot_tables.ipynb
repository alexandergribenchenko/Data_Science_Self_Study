```python
import pandas as pd

# Crear DataFrame de ejemplo
data = {'Categoria': ['A', 'B', 'A', 'B', 'A', 'B'],
        'CurrentPrice': [10.0, 20.0, 30.0, 40.0, 50.0, 60.0],
        'BaselineUnitsSold': [100, 200, 300, 400, 500, 600]}
df = pd.DataFrame(data)


df
```

```python
pivot_table = df.pivot_table(values=['CurrentPrice', 'BaselineUnitsSold'],
                            index='Categoria',
                            aggfunc={'CurrentPrice': 'sum', 'BaselineUnitsSold': 'mean'})
pivot_table
```

```python
pivot_table.columns = pivot_table.columns.map('_'.join)
tabla_dinamica = pivot_table.reset_index()
tabla_dinamica
```
