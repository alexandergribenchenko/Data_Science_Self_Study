# 02. Create Dataframe from: lists

```python
import pandas as pd
```

```python
a = [1,2,3,4,5]
b = [6,7,8,9,10]

df = pd.DataFrame(list(zip(a,b)), 
                  columns=['col_01','col_02'])
df
```

# 03. Create Dataframe from: dictionary

```python
import pandas as pd
```

```python
dict_input = { 'col_01': [1,2,3,4], 
               'col_02': ['gato','perro','gato','perro'],
               'col_03': [1,5,6,7]
             }

df = pd.DataFrame.from_dict(dict_input)
df
```

# 03. Create Dataframe from: series

```python
import pandas as pd
```

```python
author = ['Jitender', 'Purnima', 'Arpit', 'Jyoti']
article = [210, 211, 114, 178]
  
auth_series = pd.Series(author)
article_series = pd.Series(article)
  
frame = {'Author': auth_series, 'Article': article_series}
  
df = pd.DataFrame(frame)
df
```
