```python
import pandas as pd
import random
import numpy as np

random.seed(42)
np.random.seed(42)

# Generar una lista de fechas aleatorias entre junio y julio de 2023
start_date = pd.to_datetime('2023-06-01')
end_date = pd.to_datetime('2023-07-31')

date_range = pd.date_range(start_date, end_date, freq='D')
random_dates = random.sample(list(date_range), 20)

# Generar una lista de cantidades de vuelos aleatorias
flight_counts = [random.randint(0, 10) for _ in range(20)]

# Crear el DataFrame con las columnas de fecha y cantidad de vuelos
data = {'Fecha': random_dates, 'Cantidad de vuelos': flight_counts}
df = pd.DataFrame(data)

df = df.sort_values(by=['Fecha'], ascending=[True]).reset_index(drop=True)

df
```

```python
df.groupby(pd.Grouper(key='Fecha', freq='W'))['Cantidad de vuelos'].sum().reset_index()
```
