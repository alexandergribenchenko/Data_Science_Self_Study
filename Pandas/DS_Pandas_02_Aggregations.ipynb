# Pandas agreggations

```python
import pandas as pd
import numpy as np
```

```python
df_x = pd.DataFrame(np.random.rand(4,4), columns=list('abcd'))
df_x['perro'] = [0, 0, 1, 1]
df_x['gato'] = ['a', 'b', 'a', 'b']
df_x
```

```python
df_x.groupby(['perro','gato']).agg(
    min_a=('a', 'min'),
    count_a= ('a', 'count'),
    max_b=('b', 'max')
)
```
