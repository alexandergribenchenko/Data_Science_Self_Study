# 01. Filtrar columnas con regex

```python
import pandas as pd
```

```python
def regex_search(df, column, patron_regex):
    df[column] = df[column].fillna('')
    df_output = df[df[column].str.contains(patron_regex, regex=True)]
    return df_output
```

```python
data = {
    'ID': [1, 2, 3, 4, 5],
    'Textos': ['hola123', '456adios', 'texto789', 'texto', '123456']
}

df = pd.DataFrame(data)

df
```

```python
regex_search(df, 'Textos', 'adi')
```
